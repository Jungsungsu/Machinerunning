```python
import pandas as pd
```

```python
# 인공지능반 진로희망 도수에 대한 데이터 프레임을 직접 만들어보자!!
```

# 실습1

```python
df =  pd.DataFrame(data={'진로희망' : ['빅데이터분석가','자영업자','웹개발자','대학원','합계'],
                         '도수' : [6,7,4,3,20]
                         })
df
```

# 실습2

```python
# 리스트 형태로 생성
df_list = pd.DataFrame([['빅데이터분석가',6],
                        ['자영업',7],
                        ['웹개발자',4],
                        ['대학원',3],
                        ['합계',20]],
                        columns=['진로희망','도수']
                       )
df_list
```

```python
df =  pd.DataFrame(data={'진로희망' : ['빅데이터분석가','자영업자','웹개발자','대학원','합계'],
                         '도수' : [12,14,10,14,50]
                         })
df
```

```python
df_list2 = pd.DataFrame([['빅데이터분석가',12],
                        ['자영업',14],
                        ['웹개발자',10],
                        ['대학원',14],
                        ['합계',50]],
                        columns=['진로희망','도수']
                        )
df_list2
```

```python
# 두 데이터프레임 합치기 :concat
df_concat = pd.concat([df_list,df_list2['도수']],axis=1)
# 데이터 프레임의 컬럼명을 설정
df_concat.columns=['진로희망','인공지능반','빅데이터반']
df_concat
```

# 나홀로 실습

```python
# 리스트 형태로 생성
df_list = pd.DataFrame([['빅데이터분석가','6/20=0.30'],
                        ['자영업','7/20=0.35'],
                        ['웹개발자','4/20=0.20'],
                        ['대학원','3/20=0.15'],
                        ['합계','1']],
                        columns=['진로희망','상대도수']
                       )
df_list
```

```python
df_list2 = pd.DataFrame([['빅데이터분석가','12/50=0.24'],
                        ['자영업','14/50=0.28'],
                        ['웹개발자','10/50=0.20'],
                        ['대학원','14/50=0.28'],
                        ['합계','1']],
                        columns=['진로희망','상대도수']
                        )
df_list2
```

```python
# 두 데이터프레임 합치기 :concat
df_concat = pd.concat([df_list,df_list2['상대도수']],axis=1)
# 데이터 프레임의 컬럼명을 설정
df_concat.columns=['진로희망','인공지능반','빅데이터반']
df_concat
```

# 데이터 프레임 추가하기

```python
# 상대도수 데이터 프레임
df_상대도수 = pd.DataFrame([[0.3,0.24],
                            [0.35,0.28],
                            [0.2,0.2],
                            [0.15,0.27],
                            [1,1]], columns = ['인공지능반','빅데이터반']
                           )
```

```python
df_result = pd.concat([df_concat, df_상대도수], axis=1)
df_result
```

# 멀티 컬럼 설정하기

```python
# 멑티 컬럼 설정
df_result.columns = [['','도수','도수','상대도수','상대도수'],
                    ['진로희망','인공지능반','빅데이터반','인공지능반','빅데이터반']
                    ]
df_result
```
