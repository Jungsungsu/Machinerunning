```python
# numpy pandas matplotlib import 하기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine = pd.read_csv('./data/wine.csv')
wine
```

```python
# quality(1 ~ 10단계) 정답데이터로 활용
wine.columns
```

```python
# 결측치 확인
wine.info()
```

```python
# 이상치 확인
wine.describe()
```

```python
# 문제(X)와 정답(y)으로 구분
# 정답데이터 quality
X = wine.iloc[:,:-1]
y= wine.iloc[:,-1]

# 학습과 평가데이터로 구분
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                            test_size=0.3,
                                            random_state = 5)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

# C : 규제에 관한 조정
## 낮을수록 규제가 강해짐
# max_iter : 학습 횟수
## 숫자를 늘려주는게 좋음
lr_model = LogisticRegression(C = 1, max_iter = 10000)
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_model.score(X_train,y_train)
```

```python
lr_model.score(X_test, y_test)
```

```python
# GridSearch 최적의 C값 찾아보자!
from sklearn.model_selection import GridSearchCV
param = {
    'C' : [0.001,0.01,0.1,1,10,100]
}
grid_search = GridSearchCV(LogisticRegression(), param, cv=5)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.score(X_train,y_train)
```

```python
print("최적 매개변수 : " , grid_search.best_params_)
print("최고 검증 점수 : ", grid_search.best_score_)
```
